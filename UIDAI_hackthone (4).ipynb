# Heatmap For Updated biometric data

```python
# we are using ausmed data as it is taking a lot of time to get the exact value of x2,x3,x4.....
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
raw_data = {
    'Delhi': [21359000, 100.0],       
    'Haryana': [30209000, 100.0],
    'Punjab': [30730000, 100.0],
    'Kerala': [35776000, 100.0],
    'Himachal Pradesh': [7468000, 100.0],
    'Goa': [1575000, 100.0],
    'Telangana': [38090000, 100.0],
    'Andhra Pradesh': [53156000, 95.0],
    'West Bengal': [99084000, 98.6],
    'Tamil Nadu': [76860000, 97.7],
    'Karnataka': [67692000, 96.9],
    'Odisha': [46276000, 95.7],
    'Maharashtra': [126385000, 94.8],
    'Chhattisgarh': [30180000, 94.8],
    'Rajasthan': [81025000, 94.2],
    'Uttar Pradesh': [235687000, 94.0],
    'Gujarat': [71507000, 92.3],
    'Jharkhand': [39466000, 91.5],
    'Madhya Pradesh': [86579000, 91.6],
    'Assam': [35713000, 89.4],
    'Bihar': [126756000, 87.7],
    'Jammu & Kashmir': [13603000, 87.1],
    'Sikkim': [689000, 84.2],
    'Manipur': [3223000, 82.2],
    'Arunachal Pradesh': [1562000, 79.6],
    'Meghalaya': [3349000, 77.0],
    'Nagaland': [2233000, 61.4]
}

processed_data = []

for state, values in raw_data.items():
    population = values[0]
    percent_updated = values[1]
    updated_count = int(population * (percent_updated / 100))
    unupdated_count = int(population - updated_count) if percent_updated < 100 else 0
    
    processed_data.append({
        'State': state,
        'Updated Data': updated_count,
        'Pending Data': unupdated_count,
        '% Completed': percent_updated
    })

df = pd.DataFrame(processed_data)
df = df.set_index('State')

df = df.sort_values(by='% Completed', ascending=False)

heatmap_data = df[['Updated Data', 'Pending Data']]

plt.figure(figsize=(10, 14))

sns.heatmap(heatmap_data, 
            annot=True, 
            fmt=',d',   
            cmap='RdYlGn', 
            linewidths=1, 
            linecolor='white')

plt.title('State-wise Biometric Data: Updated vs Pending', fontsize=16, pad=20)
plt.ylabel('State', fontsize=12)
plt.xlabel('Dataset Status', fontsize=12)
plt.tight_layout()

plt.show()
```
